```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv('data.csv')
df
```

```python
md=df.test_score.mean()
md
```

```python
df.test_score=round(df.test_score.fillna(md),0)
df
```

```python
df.experience=df.experience.fillna(0)
df
```

```python
plt.title('MLR')
plt.xlabel('Independent Variable')
plt.ylabel('Salary')
plt.plot(df.experience,df.salary,color='red',marker='o')
plt.plot(df.test_score,df.salary,color='blue',marker='+')
plt.plot(df.interview_score,df.salary,color='skyblue',marker='*')
plt.legend(df)
plt.show()
```
